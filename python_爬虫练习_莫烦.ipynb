```python
from urllib.request  import  urlopen
import re
```

```python
html=urlopen('https://movie.douban.com/subject/26654498/').read().decode('utf-8')
```

```python
#print(html)
```

```python
res = re.findall(r'href="(.*?)"', html)
print("\nAll links: ", res)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, features='lxml')
```

```python
print(soup.p)
```

```python
all_href=soup.find_all('a')
```

```python
href=[link.get('href') for link in all_href]
href
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```
